```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("img.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap="gray")
plt.show()
```

```python
x = 60
kernel = np.ones((x, x), np.float32) / x**2
blurred = cv2.filter2D(image, -1, kernel)
plt.imshow(blurred, cmap="gray")
plt.show()
```

```python
vertical_mask = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
horizontal_mask = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

vertical_edges = cv2.filter2D(image, -1, vertical_mask)
horizontal_edges = cv2.filter2D(image, -1, horizontal_mask)
all_edges = vertical_edges + horizontal_edges
plt.imshow(all_edges, cmap="gray")
plt.show()
```
