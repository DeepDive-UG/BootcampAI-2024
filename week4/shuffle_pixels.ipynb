```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("img.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap="gray")
plt.show()
```

```python
def shuffle_image_pixels(image, seed=123):
    height, width = image.shape
    pixel_indices = np.arange(height * width)

    np.random.seed(seed)
    shuffled_indices = np.random.permutation(pixel_indices)

    flattened_image = image.flatten()
    shuffled_image = flattened_image[shuffled_indices].reshape((height, width))
    return shuffled_image
```

```python
shuffled_image = shuffle_image_pixels(image)
plt.imshow(shuffled_image, cmap="gray")
plt.show()
```
